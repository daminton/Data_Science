```python
cars1 = pd.read_csv('../csv_files/cars1.csv')  
```

```python
mpg_cylinder_wight = cars1[['mpg', 'cylinders', 'weight']]
```

```python
#https://medium.com/@sebastiannorena/finding-correlation-between-many-variables-multidimensional-dataset-with-python-5deb3f39ffb3 found this on a medium article 
corr = mpg_cylinder_wight.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(mpg_cylinder_wight.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(mpg_cylinder_wight.columns)
ax.set_yticklabels(mpg_cylinder_wight.columns)
plt.show()
```
