```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
import matplotlib.pyplot as plt
```

```python
reg = pd.read_csv('../csv_files/regression.csv') 
```

Look at the data

```python
reg.describe()
```

```python
reg.head()
```

```python
x = np.array(reg['x']).reshape((-1, 1))
y = reg['y'].to_numpy()
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x, model.predict(x), color = 'green')
plt.show()
```
