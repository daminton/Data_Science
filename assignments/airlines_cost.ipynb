```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDClassifier
```

```python
airline_data = pd.read_csv('../csv_files/airlineData.csv') 
```

```python
airline_data.head()
```

```python
length_of_flight = np.array(airline_data['Length']).reshape((-1, 1))
daily_flight_time = airline_data['FlightTime'].to_numpy()
```

```python
model = LinearRegression()
```

```python
model.fit(length_of_flight, daily_flight_time)
```

```python
plt.scatter(length_of_flight,daily_flight_time)
plt.plot(length_of_flight, model.predict(length_of_flight), color = 'green')
plt.show()
```

```python
model.score(length_of_flight, daily_flight_time)
```

```python
customers_served = np.array(airline_data['Customers']).reshape((-1, 1))
total_assets = airline_data['Assets'].to_numpy()
```
